Why might we want to keep our categorical data as 1's and 0's? Why may we want to use something like the code below to change it?<br> <br>
Keeping the 1's and 0's is easy for entering data because users are less likely to enter something incorrectly.  However, when presenting or putting this data into a viz, changing it to something more descriptive will help the audience better understand the data. 


```python
df['sex'] = df.sex.replace({1: "Male", 0: "Female"})
df['anaemia'] = df.anaemia.replace({1: "Yes", 0: "No"})
df['diabetes'] = df.diabetes.replace({1: "Yes", 0: "No"})
df['high_blood_pressure'] = df.high_blood_pressure.replace({1: "Yes", 0: "No"})
df['smoking'] = df.smoking.replace({1: "Yes", 0: "No"})

df['DEATH_EVENT'] = df.DEATH_EVENT.replace({1: "Died", 0: "Survived"})
```

## Section 2: Distribution of our data:

```python
# Plot the distribution of your variable using distplot
sam.set_theme()

sam.distplot(df['creatinine_phosphokinase'], kde = False, color ='purple', bins = 30)
```

```python
# Create boxplot to show distribution of variable
sam.boxplot(df['creatinine_phosphokinase'], color = 'green')
```

```python
# Feel free to add any additional graphs that help you answer the questions below.
sam.jointplot(x ='creatinine_phosphokinase', y ='platelets', data = df, color = 'orange')
```

```python
sam.pairplot(df, hue ="sex", palette ='Dark2')
```

```python
# Another way to check the skewness of our variable
df['creatinine_phosphokinase'].skew()
```

```python
# Another way to check the kurtosis of our variable
df['creatinine_phosphokinase'].kurtosis()
```

## Section 3: Finding Correlations

Lets start by breaking our data into two.  

```python
# splitting the dataframe into 2 parts
# on basis of DEATH_EVENT column values
df_died = df[df['DEATH_EVENT'] == 1 ]
df_lived = df[df['DEATH_EVENT'] == 0 ]
```

```python
# Plot your variable based on if they died or lived

sam.distplot(df_died['smoking'])
sam.distplot(df_lived['smoking'])
plt.title("Chances of survival vs DEATH_EVENT")
plt.legend(('Died','Lived'))
plt.plot()
```
